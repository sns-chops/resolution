# Resolution

```python
beam  =  os.path.abspath("./beam/")
instrument = 'ARCS'
Ei = 400.
```

```python
E = 0.
Q = 5
dQ_axis=(-5, 5, 0.05)
dE_axis=(-100, 100, 2.)
```

```python
ncount = 1e6
nodes = 10
```

```python
%%time
from dgsres.powder import  use_ConstantQEKernel
sim = use_ConstantQEKernel.Sim(
    instrument = instrument,
    workdir = os.path.abspath("./resolution"),
    beamdir = beam,
    Ei = Ei,
    Q = Q,
    dQ_axis=dQ_axis,
    dE_axis=dE_axis,
    ncount = ncount,
    nodes = nodes,
    )
```

```python
%%time
sim.run(E=E)
```
